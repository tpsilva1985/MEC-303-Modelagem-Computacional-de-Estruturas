```python
import numpy as np

m1, m2 = 1.0, 2.0
k1, k2, k3 = 50.0, 100.0, 50.0

M = np.diag([m1, m2])
K = np.array([[k1+k2, -k2],
              [-k2,   k2+k3]])

# Autovalores de M^{-1}K (são ω²) e autovetores (modos)
lam, V = np.linalg.eig(np.linalg.solve(M, K))
idx = np.argsort(lam)           # ordenar
lam, V = lam[idx], V[:, idx]
omega = np.sqrt(lam)
freq  = omega/(2*np.pi)

# Normalização pela massa: Phi^T M Phi = I
Phi = V.copy()
for i in range(Phi.shape[1]):
    mi = Phi[:, i].T @ M @ Phi[:, i]
    Phi[:, i] /= np.sqrt(mi)

print("ω (rad/s):", omega)
print("f (Hz):   ", freq)
print("Modos (mass-normalized):\n", Phi)
print("Check ΦᵀMΦ ≈ I:\n", Phi.T @ M @ Phi)
```
